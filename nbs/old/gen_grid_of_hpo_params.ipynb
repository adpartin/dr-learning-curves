# Create a grid of HP params using the randomized appraoch

```python
%load_ext autoreload
%autoreload 2
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')

import os
import sys
from pathlib import Path

import sklearn
import numpy as np
import pandas as pd
```

```python
seed = 0
from sklearn.model_selection import ParameterSampler
from scipy.stats.distributions import expon, uniform
import numpy as np
rng = np.random.RandomState(seed)
```

```python
# param_grid = {'a':[1, 2], 'b': expon()}

param_grid = {'opt':['sgd', 'adam'],
              'lr': uniform(0.0001, 0.001),
              'dr_rate': uniform(0.05, 0.25),
              'batch_size': [32, 64, 128, 256]}
param_list = list(ParameterSampler(param_grid, n_iter=20, random_state=rng))
# rounded_list = [dict((k, round(v, 6)) for (k, v) in d.items()) for d in param_list]
# rounded_list = [dict((k, round(v, 6)) for (k, v) in d.items()) for d in param_list]
pd.DataFrame(param_list)
```
